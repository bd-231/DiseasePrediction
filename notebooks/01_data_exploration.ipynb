# Notebook 1 — Data Exploration
Load datasets, understand structure, check missing values, duplicates, and class distribution.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

DATA = '../'
train_dis = pd.read_csv(f'{DATA}train_disease.csv')
test_dis  = pd.read_csv(f'{DATA}test_disease.csv')
drug_train = pd.read_csv(f'{DATA}UCIdrug_train.csv')
drug_test  = pd.read_csv(f'{DATA}UCIdrug_test.csv')

print('train_disease :', train_dis.shape)
print('test_disease  :', test_dis.shape)
print('drug_train    :', drug_train.shape)
print('drug_test     :', drug_test.shape)
```

## 1.1 Disease Dataset

```python
print('=== train_disease columns ===')
print(train_dis.columns.tolist())
train_dis.head(3)
```

```python
print('=== test_disease columns ===')
print(test_dis.columns.tolist())
test_dis.head(3)
```

```python
# train has an extra trailing empty column — identify it
extra_cols = set(train_dis.columns) - set(test_dis.columns)
print('Extra cols in train vs test:', extra_cols)
train_dis.dtypes.value_counts()
```

```python
print('--- Missing values (train_disease) ---')
print(train_dis.isnull().sum()[train_dis.isnull().sum() > 0])

print('\n--- Missing values (test_disease) ---')
print(test_dis.isnull().sum()[test_dis.isnull().sum() > 0])
```

```python
print('Duplicates in train_disease:', train_dis.duplicated().sum())
print('Duplicates in test_disease :', test_dis.duplicated().sum())
```

```python
print('--- Disease class distribution (train) ---')
print(train_dis['prognosis'].value_counts())
```

## 1.2 Drug Dataset

```python
drug_train.head(3)
```

```python
drug_train.dtypes
```

```python
print('--- Missing values (drug_train) ---')
print(drug_train.isnull().sum())
```

```python
print('Duplicates in drug_train:', drug_train.duplicated().sum())
print('Rating range:', drug_train['rating'].min(), '-', drug_train['rating'].max())
print('Unique drugs:', drug_train['drugName'].nunique())
print('Unique conditions:', drug_train['condition'].nunique())
```

```python
# Corrupt condition values (HTML artifacts)
bad_cond = drug_train['condition'].str.contains('users found', na=False)
print('Corrupt condition rows:', bad_cond.sum())
drug_train[bad_cond]['condition'].head()
```

```python
print('\nTop 20 drug conditions:')
clean = drug_train[~bad_cond]
print(clean['condition'].value_counts().head(20))
```
